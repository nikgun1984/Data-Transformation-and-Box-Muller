```python
import numpy as np
import matplotlib.pyplot as plt
import timeit

start = timeit.default_timer()
#generate from uniform dist
np.random.seed()
N = 100000000
z1 = np.random.uniform(0, 1.0 ,N)
z2 = np.random.uniform(0, 1.0 ,N)

#transformation of Box-Muller
#y1 = z1*((-2*np.log(r**2))/r**2)**(0.5)
y1 = np.sqrt(-2*np.log(z1))*np.cos(2*np.pi*z2)
#y2 = z2*((-2*np.log(r**2))/r**2)**(0.5)
y2 = np.sqrt(-2*np.log(z1))*np.sin(2*np.pi*z2)

ghosts = y1[y1 >= 5]
ghosts = y1[y1 <= -5]
ghosts1 = y2[y2 >= 5]
ghosts1 = y2[y2 <= -5]
print(ghosts)
count = ghosts.size
count1 = ghosts1.size
print("The number of outliers or \"ghost\" values is %s" %(count))
print("The number of outliers or \"ghost\" values is %s" %(count1))

#discard outlier
y1 = y1[y1 <= 5]
y1 = y1[y1 >= -5]
y2 = y2[y2 <= 5]
y2 = y2[y2 >= -5]
#plot
fig = plt.figure()
ax = fig.add_subplot(2,1,1)
ax.hist(y1,bins=50,color='red')
plt.title("Histogram")
plt.xlabel("y1")
plt.ylabel("frequency")
ax2 = fig.add_subplot(2,1,2)
ax2.hist(y2,bins=50,color='blue')
plt.xlabel("y2")
plt.ylabel("frequency")
plt.show()

stop = timeit.default_timer()
print('Time: ', stop - start)  
```
